```python
path = "/content/gdrive/MyDrive/ML-Transportation-Rate/zipcode"
us_zip = pd.read_excel(path +'/us_city.xls')
canada_zip = pd.read_csv(path + '/canada_zip.csv')
```

```python
# data cleaning for us_zip dataset
us_zip['State'] = [str(i)[-3:-1] for i in us_zip['State']]
us_zip = us_zip[['Zip', 'State', 'Longitude','Latitude']]
us_zip.columns = ['zipcode','state','lng','lat']
us_zip['zipcode'] = ["0" * (5 - len(str(i))) + str(i) for i in us_zip['zipcode']]

# data cleaning for canada_zip dataset
canada_zip = canada_zip[['zipCode', 'stateISO', 'latitude', 'longitude']]
canada_zip.columns = ['zipcode', 'state', 'lat','lng']
canada_columns_order = ['zipcode', 'state', 'lng','lat']
canada_zip = canada_zip[canada_columns_order]
canada_zip = canada_zip[canada_zip['zipcode'].str.len() == 6]

zip = us_zip.append(canada_zip)
```

```python
zip['lng'] = [i*[1,-1][i>0] for i in zip['lng']]
```

```python
zip = zip.groupby(['zipcode', 'state']).agg({'lng':'mean', 'lat':'mean'})
zip = zip.reset_index()
```
